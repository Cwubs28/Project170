```python
import pandas as pd
import lyricsgenius

# Reading file to get access token for Genius API
with open("access_token", "r") as file:
    access_token = file.read()

# Creating genius object
genius = lyricsgenius.Genius(access_token)

#File of song URls on Genius top charts
songs = "songs.txt"
####artist = genius.search_artist("Drake", max_songs = 1, sort = "popularity")

# Using function to get data structure (Key: song_url, Value: file name that stores the lyrics
data = process_songs(songs)

print(data)

```

```python

def process_songs(file_name):
    song_data = {}
    
    with open(file_name, "r") as file:
        for song_num, url in enumerate (file, start = 1):
            url = url.strip()
            song_lyrics = genius.lyrics(song_url = url, remove_section_headers = False)
            #name = input("What would you like to name the txt file for " + url)

            song_name = "song" + str(song_num)
            with open(song_name, "w") as file:
                song_file = file.write(song_lyrics)
            song_data[url] = song_name

    return song_data
            
    
```
